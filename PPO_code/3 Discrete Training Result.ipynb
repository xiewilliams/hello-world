```python
import gym
import numpy as np
import matplotlib.pyplot as plt
```

```python
from train_ram_discrete import train
from agent import Agent_discrete as Agent
```

```python
env = gym.make('LunarLander-v2')
state_size, action_size = env.observation_space.shape[0], env.action_space.n
state_size, action_size
```

## Default Setting, $\beta=0, \epsilon=0.2, \tau=0.99$

```python
agent = Agent(state_size, action_size, lr=0.005, beta=0, eps=0.2, tau=0.99, gamma=0.99, device=device, mode='TD', share=False, use_critic=True, normalize=False)
rewards, average_log = train(agent, env, n_episode=2000, n_update=10, update_frequency=4, max_t=1000, scale=1)

plt.figure(figsize=(17, 8))
plt.plot(rewards, label='episodic reward', color='black')
plt.plot(average_log, label='moving average', color='green')
plt.hlines(y=200, xmin=0, xmax=2000, label='baseline', color='blue')
plt.title('Share={}, use_critic={}, normalize={}, lr={}, Scale={}'.format(agent.share, agent.use_critic, agent.normalize, agent.lr, 1))
plt.legend()
plt.show()
```
